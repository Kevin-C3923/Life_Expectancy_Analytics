```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(Path("merged_pop_df.csv"))
df.head(61)
```

```python
# update to not include features with Zero
df.groupby(['County']).mean(0)
```

```python
df = df.rename(columns = {'Unnamed: 0':'County_Number'})
#clean_data.rename(columns = {'% Fair/Poor':'% Fair or Poor Health',
```

```python
df.dtypes
```

```python
df_w_bronx= df.drop(labels =2,axis=0)
df_w_bronx
```

```python
plt.scatter(df_w_bronx.Fair_or_Poor_Health_Percent,df_w_bronx.Death_Ratio)
plt.xlabel('Fair or Poor Health')
plt.ylabel('Premature Deaths')
plt.show()
```

```python
X = df_w_bronx.Fair_or_Poor_Health_Percent.values.reshape(-1, 1)
```

```python
X[:5]
```

```python
X.shape
```

```python
y = df_w_bronx.Death_Ratio
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
y_pred = model.predict(X)
print(y_pred.shape)
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.show()
```

```python
print(model.coef_)
print(model.intercept_)
```
