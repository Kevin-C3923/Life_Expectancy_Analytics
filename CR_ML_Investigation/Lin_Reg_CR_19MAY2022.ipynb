```python
populations_df = pd.read_csv(Path('2020_Census_Population.csv'))
populations_df.head(61)
```

```python
#merged_pop_df = pd.merge(df, populations_df, on=["County", "population"])
merged_pop_df = pd.merge(means_df, populations_df, on=["County"])


merged_pop_df.shape
```

```python
merged_pop_df.head(200)
```

```python
#Create deaths as raio.

merged_pop_df.dtypes
```

```python
merged_pop_df['population'] = merged_pop_df['population'].astype(float)
```

```python
merged_pop_df.dtypes
```

```python
merged_pop_df["Death_Ratio"]=merged_pop_df["Deaths"]/merged_pop_df["population"]
merged_pop_df["Driving_Death_Ratio"]=merged_pop_df["Driving_Deaths_Number"]/merged_pop_df["population"]
merged_pop_df["Uninsured_Ratio"]=merged_pop_df["Uninsured_Number"]/merged_pop_df["population"]



```

```python
merged_pop_df.head()
```

```python
#clean up reformatted data
merged_pop_df.drop(["Deaths", "Driving_Deaths_Number", "Uninsured_Number", "population"], axis=1, inplace=True)

#remove unnecessary additional data
merged_pop_df.drop(["Years","FIPS"], axis=1, inplace=True)

```
