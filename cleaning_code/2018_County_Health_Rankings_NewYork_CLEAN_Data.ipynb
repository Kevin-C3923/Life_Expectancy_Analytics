```python
import pandas as pd
```

```python
# Reading in Data
df = pd.read_csv('csv_data/2018_County_Health_Rankings_New_York_Data.csv')
df
```

```python
#Select the columns that we want
select_df = df[['FIPS', 'County','% Fair/Poor','% LBW','Food Environment Index','# Driving Deaths','# Uninsured','MHP Ratio']]
select_df
```

```python
# Removing the first row that was NaN 
select_df = select_df.drop(labels=0, axis=0)
select_df
```

```python
# The the year for the dataset
select_df["Years"] = "2018"
select_df
```

```python
# Reorganizing the dataset
clean_data = select_df[['Years','FIPS', 'County','% Fair/Poor','% LBW','Food Environment Index','# Driving Deaths','# Uninsured','MHP Ratio']]
clean_data
```

```python
# Spliting the Mental Health Provider Ratio 
changing = clean_data['MHP Ratio'].str.split(':', expand=True)
changing.dtypes
```

```python
# Converting the string into int
changing = changing.astype('int')
```

```python
# Changing the Mental Health Provider Ratio into a int value
clean_data['MHP Ratio'] = changing[1]/changing[0]
clean_data
```

```python
clean_data.shape
```

```python
#convert all objects to integer
df = pd.DataFrame(clean_data)
df = df.astype({"FIPS":"int",
                        "Years":"int",
                        "% Fair/Poor":"int",
                        "% LBW":"int",
                        "# Driving Deaths":"int",
                        "# Uninsured":"int"})
```

```python
# Saving the new and cleaned dataset
file_path = "2018_Clearned_County_Health_Rankings_New_York_Data.csv"
clean_data.to_csv(file_path, index=False )
```
