
## 2021 Data Cleaning 


```python
import pandas as pd
# Reading in Data
df_2021 = pd.read_csv('csv_data/2021_County_Health_Rankings_New_York_Data.csv')
df_2021
```

```python
# Selecting the columns we only want
select_df_2021 = df_2021[['FIPS', 'County', 'Deaths', '% Fair or Poor Health', '% Low birthweight', 
            'Food Environment Index', '# Driving Deaths', '# Uninsured', 'Mental Health Provider Ratio']]
select_df_2021
```

```python
# Removing the first row that was NaN 
select_df_2021 = select_df_2021.drop(labels=0, axis=0)
select_df_2021
```

```python
# The the year for the dataset
select_df_2021["Years"] = "2021"
select_df_2021
```

```python
# Reorganizing the dataset
clean_data_2021 = select_df_2021[['Years', 'FIPS', 'County', 'Deaths', '% Fair or Poor Health', '% Low birthweight', 
            'Food Environment Index', '# Driving Deaths', '# Uninsured', 'Mental Health Provider Ratio']]
clean_data_2021
```

```python
clean_data_2021.dtypes
```

```python
# Spliting the Mental Health Provider Ratio 
changing_2021 = clean_data_2021['Mental Health Provider Ratio'].str.split(':', expand=True)
changing_2021.dtypes
```

```python
# Converting the string into int
changing_2021 = changing_2021.astype('int')
```

```python
# Changing the Mental Health Provider Ratio into a int value
clean_data_2021['Mental Health Provider Ratio'] = changing_2021[1]/changing_2021[0]
clean_data_2021
```

```python
# Saving the new and cleaned dataset
file_path_2021 = "csv_data/2021_clean_dataset.csv"
clean_data_2021.to_csv(file_path_2021, index=False )
```
